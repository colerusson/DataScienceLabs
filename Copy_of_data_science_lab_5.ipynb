## Exercise 1: Protein Powder
Carbs, fats and proteins are the three primary macro nutrients. Create a figure plotting the distribution of each of these macro nutrients together (i.e., three distributions on a single plot). Make sure to provide a legend.

```python
# Enter all of your code for exercise 1 here. Feel free to add more cells if you need to:
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('cereal.csv')

plt.hist(df['carbo'], bins=20, alpha=0.5, label='Carbs')
plt.hist(df['fat'], bins=20, alpha=0.5, label='Fats')
plt.hist(df['protein'], bins=20, alpha=0.5, label='Proteins')

plt.legend()
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Macro Nutrients')

plt.show()
```

## Exercise 5: America Part 2
Make five plots comparing the relationships of carbo, sugars, calories, protein, and fat with rating.

```python
# Write your code to compare the various variables with rating below:
df = pd.read_csv('cereal.csv')

sns.regplot(x='carbo', y='rating', data=df)
plt.show()
correlation_carbo = df['carbo'].corr(df['rating'])
print(correlation_carbo)

sns.regplot(x='sugars', y='rating', data=df)
plt.show()
correlation_sugar = df['sugars'].corr(df['rating'])
print(correlation_sugar)

sns.regplot(x='calories', y='rating', data=df)
plt.show()
correlation_cal = df['calories'].corr(df['rating'])
print(correlation_cal)

sns.regplot(x='protein', y='rating', data=df)
plt.show()
correlation_protein = df['protein'].corr(df['rating'])
print(correlation_protein)

sns.regplot(x='fat', y='rating', data=df)
plt.show()
correlation_fat = df['fat'].corr(df['rating'])
print(correlation_fat)
```

Of the variables carbo, sugars, calories, protein, and fat, which has the strongest relationship with rating? Justify your answer.

Of the 5 variables, sugar has the strongest relationship with rating, as it has the strongest r value correlation (closest to -1 or 1), and the graph also visually shows that it has the most accurate line of best fit with the least room of variance.

## Exercise 7: Its Hot and It's Cold
The type column has two values: H='hot' and C='cold'. What is the average rating of each type?

```python
# Write your code to get the average rating by type of cereal below:
df = pd.read_csv('cereal.csv')

types = df.groupby('type')
averageRating = types['rating'].mean()
print(averageRating)
```
