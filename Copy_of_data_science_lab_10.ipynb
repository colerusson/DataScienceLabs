After reading the above documentation (You can skim it) attempt to explain to me like I'm five, what these algorithms are doing:
 

Explain how K-Means works?

K-Means is an algorithm that tries to group data points into a certain number of clusters, with each cluster having similar features. It chooses a number of clusters to create. It randomly assigns each data point to one of the clusters. Then it calculates the center of each cluster. It reassigns each data point to the closest center cluster. It repeats that until there is no more assignment change or until a fixed set of iterations.

Explain how DBScan works?

DBSCAN is an algorithm that groups data points into clusters based on their density. It stars with a random data point and checks for nearby data points. If there are enough nearby, those points become a cluster. This repeats for each of the nearby points, and then for each of their nearby points, until no more nearby points. Then this repeats form the beginning with a new random, unassigned point. 

Explain how GMM works?

GMM stands for Gaussian Mixture Model, which is a statistical model that assumes that the data is generated by a mixture of several Gaussian distributions. It randomly assigns each point to a cluster. It calculates the mean and variance of each cluster, which are the center and spread of the distribution. Using those values, it calculates the probability that each data point belongs to each cluster. Then it assigns the poitn to the cluster with the highest probability. Then it repeats until there is no more assignment change or until fixed iterations.

#### Datasets:
We will generate three toy datasets using the scikit-learn api, which you can do with the following code:

```python
from sklearn import datasets
np.random.seed(0)

# ============
# Generate datasets. We choose the size big enough to see the scalability
# of the algorithms, but not too big to avoid too long running times
# ============
n_samples = 1500
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=0.5, noise=0.05)[0]
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=0.05)[0]
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)[0]
```

This generates three datasets, `noisy_circles`, `noisy_moons`, and `blobs`

Your Task: 
* We want you to train each clustering algorithm on each dataset (you should have a total of 9 plots).
* For each dataset/algorithm, plot the points. Color the points using the cluster the belong to

```python
#Enter the code for exercise 1 here
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
np.random.seed(0)

kmeans = KMeans(n_clusters=2)
dbscan = DBSCAN(eps=0.3, min_samples=5)
agg = AgglomerativeClustering(n_clusters=2)

datasets = [(noisy_circles, 'Noisy Circles'), (noisy_moons, 'Noisy Moons'), (blobs, 'Blobs')]
algorithms = [(kmeans, 'KMeans'), (dbscan, 'DBSCAN'), (agg, 'Agglomerative Clustering')]

for dataset, dataset_name in datasets:
    X = dataset
    plt.figure(figsize=(15, 5))
    
    for algorithm, algorithm_name in algorithms:
        algorithm.fit(X)
        y_pred = algorithm.labels_.astype(np.int)
        plt.subplot(1, 3, algorithms.index((algorithm, algorithm_name)) + 1)
        plt.scatter(X[:, 0], X[:, 1], c=y_pred)
        plt.title(f"{algorithm_name} on {dataset_name}")
        
    plt.show()
```

## Exercise 4: K Nearest Neighbors

Your Task: 
* Train a K-nearest neighbors (sklearn.neighbors.KNeighborsClassifier ) on the iris data.

* Train your KNN when the n_neighbors parameter is 5. Report your train accuracy and test accuracy

* Perform a grid search over the parameter n_neighbors over the range 1-20:

* For each value of n_neighbors, fit a KNN and record your train and test accuracy

* Create a plot showing the test/train accuracy over the n_neighbors



```python
# Enter your code for Exercise 4 here
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Load iris dataset
df = sm.datasets.get_rdataset(dataname='iris', package='datasets').data
X = df.iloc[:, :4].values
y = df.iloc[:, 4].values

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train a KNN classifier with n_neighbors=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Print train and test accuracy
train_acc = knn.score(X_train, y_train)
test_acc = knn.score(X_test, y_test)
print("Train accuracy: ", train_acc)
print("Test accuracy: ", test_acc)

# Grid search over n_neighbors
param_grid = {'n_neighbors': range(1, 21)}
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, return_train_score=True)
grid_search.fit(X_train, y_train)

# Print train and test accuracy for each value of n_neighbors
train_scores = []
test_scores = []
for i in range(len(param_grid['n_neighbors'])):
    train_scores.append(grid_search.cv_results_['mean_train_score'][i])
    test_scores.append(grid_search.cv_results_['mean_test_score'][i])
    print("n_neighbors = {}: Train accuracy = {:.3f}, Test accuracy = {:.3f}".format(
        param_grid['n_neighbors'][i], train_scores[-1], test_scores[-1]))

# Plot train and test accuracy as a function of n_neighbors
plt.plot(param_grid['n_neighbors'], train_scores, label='Train')
plt.plot(param_grid['n_neighbors'], test_scores, label='Test')
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

What is your train accuracy and test accuracy for when the n_neighbors parameter is 5

n_neighbors = 5: Train accuracy = 0.963, Test accuracy = 0.933
